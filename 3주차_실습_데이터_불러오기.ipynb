# **정규화 (Normalization) 공식 정리**

정규화(Normalization)는 데이터의 범위를 일정한 스케일로 변환하여 **분석, 머신러닝 모델 학습, 데이터 비교** 등을 용이하게 하는 기법입니다. 가장 일반적인 정규화 방법과 그 수식을 정리하면 다음과 같습니다.

---

## **1. Min-Max 정규화 (Min-Max Normalization)**
- 데이터 값을 **\[0,1\] 또는 \[-1,1\] 범위**로 변환하는 방법  
- 데이터의 상대적인 크기를 유지하면서 특정 범위로 조정할 때 사용  

```math
X_{\text{norm}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}
```

### **설명**
- \(X\): 원본 데이터 값  
- \(X_{\min}\): 데이터의 최소값  
- \(X_{\max}\): 데이터의 최대값  
- \(X_{\text{norm}}\): 정규화된 데이터 값 (0~1 범위)

### **예시**
| 원본 데이터 (\(X\)) | 최소값 (\(X_{\min}\)) | 최대값 (\(X_{\max}\)) | 정규화된 값 (\(X_{\text{norm}}\)) |
|------|------|------|------|
| 20 | 10 | 50 | \(\frac{20-10}{50-10} = 0.25\) |
| 30 | 10 | 50 | \(\frac{30-10}{50-10} = 0.5\) |
| 50 | 10 | 50 | \(\frac{50-10}{50-10} = 1.0\) |

---

## **2. Z-Score 정규화 (Standardization)**
- 데이터의 평균을 **0**, 표준 편차를 **1**로 맞추는 방법  
- 데이터의 분포를 유지하면서 표준 정규 분포(평균=0, 표준편차=1)로 변환할 때 사용  

```math
X_{\text{norm}} = \frac{X - \mu}{\sigma}
```

### **설명**
- \(X\): 원본 데이터 값  
- \(\mu\): 데이터의 평균  
- \(\sigma\): 데이터의 표준 편차  
- \(X_{\text{norm}}\): 정규화된 데이터 값 (평균 0, 표준 편차 1)

### **예시**
| 원본 데이터 (\(X\)) | 평균 (\(\mu\)) | 표준 편차 (\(\sigma\)) | 정규화된 값 (\(X_{\text{norm}}\)) |
|------|------|------|------|
| 20 | 30 | 10 | \(\frac{20-30}{10} = -1.0\) |
| 30 | 30 | 10 | \(\frac{30-30}{10} = 0.0\) |
| 40 | 30 | 10 | \(\frac{40-30}{10} = 1.0\) |

---

## **3. 최대 절댓값 정규화 (Max Abs Normalization)**
- 데이터의 최댓값을 기준으로 **-1에서 1 사이**로 변환  
- 음수 값이 포함된 경우에도 **원본 데이터의 방향성을 유지**할 때 유용  

```math
X_{\text{norm}} = \frac{X}{|X_{\max}|}
```

### **설명**
- \(X\): 원본 데이터 값  
- \(X_{\max}\): 데이터 절댓값 중 최대값  
- \(X_{\text{norm}}\): 정규화된 데이터 값 (-1~1 범위)

### **예시**
| 원본 데이터 (\(X\)) | 최대 절댓값 (\(|X_{\max}|\)) | 정규화된 값 (\(X_{\text{norm}}\)) |
|------|------|------|
| -50 | 100 | \(\frac{-50}{100} = -0.5\) |
| 25 | 100 | \(\frac{25}{100} = 0.25\) |
| 100 | 100 | \(\frac{100}{100} = 1.0\) |

---

## **4. 로그 정규화 (Log Normalization)**
- 데이터 값의 크기 차이가 클 때 **로그 변환**을 적용하여 분포를 조정  
- 특히 **지수적으로 증가하는 데이터**(예: 소득, 인구 수)에서 많이 사용  

```math
X_{\text{norm}} = \log(X + 1)
```

### **설명**
- \(X\): 원본 데이터 값  
- 로그 변환 시 0 이하의 값이 존재하면 +1을 더해 변환  
- 값이 클수록 상대적인 크기가 줄어들어 스케일 조정 효과 발생

### **예시**
| 원본 데이터 (\(X\)) | 정규화된 값 (\(X_{\text{norm}}\)) |
|------|------|
| 1 | \(\log(1+1) = 0.693\) |
| 10 | \(\log(10+1) = 2.398\) |
| 100 | \(\log(100+1) = 4.615\) |

---

## **5. 정규화 선택 기준**
| 정규화 방법 | 사용 상황 |
|------------|--------------------------------|
| **Min-Max 정규화** | 데이터 범위를 일정한 구간으로 변환할 때 |
| **Z-Score 정규화** | 평균과 표준 편차를 기준으로 정규화할 때 |
| **Max Abs 정규화** | 음수 데이터를 포함한 경우, 방향성을 유지할 때 |
| **로그 정규화** | 데이터의 분포가 기하급수적으로 증가할 때 |

---
